```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# ğŸ“‚ Define the saving folder
base_path =  "..."
save_path = os.path.join(base_path, str(listing_id))
os.makedirs(save_path, exist_ok=True)

# ğŸ”¹ Modern style configuration
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))

# ğŸ¨ **Define colors**
color_box = "#4682B4"     # Elegant blue for boxplot
color_listing = "#D72638" # Red for the current listing
color_bg = "#F8F9FA"      # Light gray background

# ğŸ“Œ Apply background to the chart
plt.gca().set_facecolor(color_bg)

# ğŸ  **Filter only the listings with the same number of rooms as the current listing**
habitaciones_listing = listing['habitaciones'].values[0]
df_filtrado = df[df['habitaciones'] == habitaciones_listing].copy()

# ğŸ“Š **Boxplot of â‚¬/mÂ² for the number of rooms of the current listing (axes transposed)**
ax = sns.boxplot(
    data=df_filtrado, y='habitaciones', x='â‚¬/mÂ²', 
    color=color_box, width=0.4, linewidth=1.5,
    boxprops=dict(facecolor=color_box, alpha=0.5, edgecolor="black"),
    medianprops=dict(color="white", linewidth=2),  # White line for the median
    whiskerprops=dict(color="black", linewidth=1.5),
    capprops=dict(color="black", linewidth=1.5),
    orient="h"  # <-- Transpose axes
)

# ğŸ”¥ **Adjust axes to include all data**
x_min = df_filtrado['â‚¬/mÂ²'].min() * 0.9  # 10% below the minimum
x_max = max(df_filtrado['â‚¬/mÂ²'].max(), listing['â‚¬/mÂ²'].values[0]) * 1.1  # 10% above the maximum
plt.xlim(x_min, x_max)  # Ensure both the red point and boxplot are visible

# ğŸ”´ **Highlight the current listing (aligned to the center of the category)**
plt.scatter(
    x=[listing['â‚¬/mÂ²'].values[0]], y=[0],  # y=0 to center in the only category
    color=color_listing, s=150, marker="D", edgecolor="black", linewidth=2, label="Current Listing"
)

# ğŸ“ **Annotation for the current listing**
plt.annotate(
    f"{listing['â‚¬/mÂ²'].values[0]:,.0f} â‚¬/mÂ²", 
    xy=(listing['â‚¬/mÂ²'].values[0], 0),
    xytext=(listing['â‚¬/mÂ²'].values[0] + (x_max - x_min) * 0.05, 0.1),  # Adjust text position
    arrowprops=dict(arrowstyle="->", color=color_listing, lw=1.5),
    fontsize=12, fontweight="bold", color=color_listing,
    ha="left"
)

# ğŸ¢ **Labels and Title**
plt.ylabel("Number of Rooms", fontsize=14, fontweight="bold", color="#333333")  # ğŸ”„ Now on Y-axis
plt.xlabel("â‚¬/mÂ²", fontsize=14, fontweight="bold", color="#333333")            # ğŸ”„ Now on X-axis
plt.title(f"ğŸ“Š â‚¬/mÂ² in properties with {habitaciones_listing} rooms", fontsize=16, fontweight="bold", color="#222222")

# ğŸ¯ **Legend**
plt.legend(fontsize=12, frameon=True, facecolor="white", edgecolor="gray", loc="upper right")

# ğŸ›  **Final adjustments**
sns.despine(left=True, bottom=True)
plt.yticks([0], [habitaciones_listing], color="#444444", fontsize=12)  # Ensure only one label on Y-axis
plt.xticks(color="#444444", fontsize=12)

# ğŸ“‚ Save the chart in the property folder
file_path = os.path.join(save_path, f"precio_por_habitaciones_{listing_id}.png")
plt.savefig(file_path, dpi=300, bbox_inches="tight")

# ğŸ“¢ Confirmation
print(f"âœ… Chart saved at: {file_path}")

# ğŸ“Š **Show chart**
plt.show()
```
